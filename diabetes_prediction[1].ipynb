```python
#importing libraries and loading datasets
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as p
import numpy as np
from sklearn.model_selection import train_test_split
X,y=load_diabetes(return_X_y=True)
print(X,y)
p.plot(X,y,marker='o',linestyle='none')
p.show()
```

```python
#training and testing the dataset
xtrain,x_test,ytrain,y_test=train_test_split(X,y,test_size=0.1,random_state=50)
print(xtrain)
print(ytrain)
print(x_test)
print(y_test)

```

```python
#performing linear regression using scikit learn
from sklearn.linear_model import LinearRegression
learn=LinearRegression()
learn.fit(xtrain,ytrain)
y1=learn.predict(x_test)
print("coeffiecent:",learn.coef_)
print("intercept:",learn.intercept_)
p.scatter(y_test,y1,color='r')
p.plot(y1,y1)
p.show()
```

```python
#calculating the mean squared error
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y1))
```
